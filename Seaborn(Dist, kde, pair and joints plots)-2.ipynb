```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
```

```python
# Problem 1: Plot a rugplot on given below random generated data.
num = np.random.randn(170)
plt.figure(figsize=(15,10))
sns.distplot(num,hist=False,kde=False,rug=True)
plt.show()
```

```python
# Problem 2: Plot a pairs plot on iris data set showing different species. Write your conclusions as well.
iris = sns.load_dataset('iris')
sns.pairplot(iris,hue='species',aspect=1.2);
```

## Conclusions :
- Species Setosa has the clearest distinction from the other two species( petal_lenght and petal_width are clearly defined )
- There is a high positive correlation between Petal_Length and Petal_Width for all 3 species
- There is high overlapping between the species Versicolor and Virginica( particularly for Setal_Length and Setal_Width )
- Verigina has the largest features and Setosa has the smallest, except Sepal_Width
- Sepal_Width is almost similar for all 3 species

```python
# Problem 3: Plot a joint plot showing relation between 'petal length' and 'petal width' with density estimates.
sns.jointplot(x='petal_length',y='petal_width',kind='kde',data=iris,height=15);
```

```python
# Problem 4: Plot a pie chart of survived based on sex.
titanic=sns.load_dataset('titanic')
plot=titanic[titanic['survived']==1]['sex'].value_counts()
plt.figure(figsize=(15,15))
plt.pie(plot.values,labels=plot.index,autopct='%1.2f%%',explode=[0.11,0],shadow=True, startangle=90)
plt.show()
```
